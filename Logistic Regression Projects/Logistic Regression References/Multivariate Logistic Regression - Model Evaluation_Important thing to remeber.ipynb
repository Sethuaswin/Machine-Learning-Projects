## <font color='blue'>Important things to remember

**Confusion Matrix:**

|Actual/Predicted|No|Yes|
|:--|:--|:--|
|No|TN|FP|
|Yes|FN|TP|

TN - True Negatives<br>
FP - False Positives<br>
FN - False Negatives<br>
TP - True Positives<br>


- **Precision:** Probability that a predicted `Yes` is actually a `Yes`.

<center>Precision Elements in a Confusion Matrix


The formula for precision can be given as:

$$Precision = \frac{TP}{TP+FP}$$

Remember that `Precision` is the same as the `Positive Predictive Value` that you learnt about earlier. From now on, we will call it precision.

<br><br>
- **Recall:**  Probability that an actual `Yes` case is predicted correctly.

<div>
<img src="https://cdn.upgrad.com/UpGrad/temp/f96174e7-6f21-4ae8-a990-65b53427d917/Recall.PNG" width='250'>
</div>
<center>Recall Elements in a Confusion Matrix
    
The formula for recall can be given as:
    
$$Recall = \frac{TP}{TP+FN}$$
    
Remember that `Recall` is exactly the same as `sensitivity`. Don't get confused between these.
    
You might be wondering, if these are almost the same, then why even study them separately? The main reason behind this is that in the industry, some businesses follow the `Sensitivity-Specificity` view and some other businesses follow the `Precision-Recall` view and hence, will be helpful for you if you know both these standard pairs of metrics.

#### <font color='blue'>Understanding ROC Curve (receiver operating characteristic curve)

The ROC Curve Being plotted between False Positive Rate(FPR) in x-axis, True Positive Rate (TPR) in y-axis, to evluate the model higher the **Area Under The ROC curve is better**

Lets see the below example



**True Positive Rate (TPR)**:

This value gives you the number of positives correctly predicted divided by the total number of positives.Its formula as shown below,


$$True \ Positive \ Rate(TPR) = \frac{True \ Positive}{Total \ Number \ of \ Actual \ Positives}$$

Now, recall the labels in the confusion matrix,

<img src='https://cdn.upgrad.com/UpGrad/temp/eb23ecb5-460a-4246-bf5b-01e9914b1a96/Confusion+ROC.PNG' width='300'>

As you can see, the highlighted portion shows the row containing the total number of actual positives. Therefore, the denominator term, i.e. in the formula for TPR is nothing but -

$$Total \ Number \ of \ Actual \ Positives = True \ Positives + False \ Negatives$$

So, the formula for True Positive Rate (TPR) becomes -

$$True \ Positive \ Rate(TPR) = \frac{True \ Positive}{True \ Positives+False \ Negatives} = \frac{TP}{TP+FN}$$ 


As we might remember, the above formula is nothing but the formula for sensitivity. Hence, the term True Positive Rate that you just learnt about is nothing but sensitivity.

The second term which you saw was -

**False Positive Rate (FPR)**:

This term gives you the number of false positives (0s predicted as 1s) divided by the total number of negatives. The formula is -

$$False \ Positive \ Rate(FPR) = \frac{False \ Positive}{True \ Negatives+False \ Positives} = \frac{FP}{TN+FP}$$ 

Again, if you recall the formula for specificity, it is given by - 

$$Specificity = \frac{TN}{TN+FP}$$

Hence, you can see that the formula for False Positive Rate (FPR) is nothing but **(1 - Specificity)**. You can easily verify it yourself.

**Formula**

|Parameter|Formula|
|:----|:----|
|Accuracy|$$\frac{Correctly \ Predict \ Lables}{Total \ Number \ of \ Labels}=\frac{TN+TP}{TN+FP+FN+TP}$$|
|Sensitivity|$$\frac{Number  \ of \ actual \ Yeses \ correctly \ predicted}{Total \ number \ of \ actual \ Yeses}=\frac{TP}{TP+FN}$$|
|Specificity|$$\frac{Number  \ of \ actual \ Nos \ correctly \ predicted}{Total \ number \ of \ actual \ No}=\frac{TN}{TN+FP}$$|
|True Positive Rate(TPR)|$$\frac{True \ Positive}{True \ Positives+False \ Negatives} = \frac{TP}{TP+FN}$$|
|True Positive Rate(TPR)|$$True Positive Rate = Sensitivity$$|
|False Positive Rate(FPR)|$$\frac{False \ Positive}{True \ Negatives+False \ Positives} = \frac{FP}{TN+FP}$$|
|False Positive Rate(FPR)|$$(1-Specificity)$$|
|Positive Predicted Value|$$\frac{TP}{TP+FP}$$|
|Negative Predicted Value|$$\frac{TN}{TN+FN}$$|
|Precision|$$\frac{TP}{TP+FP}$$|
|Precision|$$Precision = Positive \ Predicted \ Value$$|
|Recall|$$\frac{TP}{TP+FN}$$|
|Recall|$$Recall = True \ Positive \ Rate = Sensitivity$$|
|F1-Score|$$2\times{\frac{precision\times{recall}}{precision+recall}}$$|
